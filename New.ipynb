## Importing Libraries

## Importing Datasets


```python
content = pd.read_csv("./Dataset/Content.csv")
location = pd.read_csv("./Dataset/Location.csv")
profile = pd.read_csv("./Dataset/Profile.csv")
reactions = pd.read_csv("./Dataset/Reactions.csv")
reaction_types = pd.read_csv("./Dataset/ReactionTypes.csv")
session = pd.read_csv("./Dataset/Session.csv")
user = pd.read_csv("./Dataset/User.csv")
```

## Cleaning of Datasets


```python
# Drop "Unnamed: 0" column for datasets

content.drop("Unnamed: 0", axis=1, inplace=True)
location.drop("Unnamed: 0", axis=1, inplace=True)
reactions.drop(["Unnamed: 0", "User ID"], axis=1, inplace=True)
reaction_types.drop("Unnamed: 0", axis=1, inplace=True)
profile.drop("Unnamed: 0", axis=1, inplace=True)
session.drop("Unnamed: 0", axis=1, inplace=True)
```

## Combining all the Datasets


```python
# Merging and Renaming Columns

df = content.merge(profile, how="left", on="User ID")
df = df.merge(session, how="left", on="User ID")
df = df.merge(reactions, how="left", on="Content ID")
df = df.rename(columns={"Type_x": "Cont_Type", "Type_y": "Type"})
df = df.merge(reaction_types, how="left", on="Type")
df = df.rename(columns={"Type": "React_Type"})
```

## Dropping columns which are not required


```python
# Dropping "URL" and "Interests" columns

df.drop(["URL", "Interests"], axis=1, inplace=True)
```

```python
df.head()
```

```python
df.dropna()
```

```python
df["Category"].unique()
```

```python
# Converting Category in lowercase and removing ""

df["Category"] = df["Category"].replace('"', "", regex=True)
df["Category"] = df["Category"].str.lower()
```

```python
df["Category"].unique()
```

## Exporting to .csv


```python
df.to_csv("Data.csv", index=False)
```

```python
top_reactions = (
    df.groupby("Category")
    .agg({"React_Type": "count"})
    .reset_index()
    .sort_values(by="React_Type", ascending=False)
    .head(5)
)

top_reactions
```

```python
top_categories = (
    df.groupby("Category")
    .agg({"Score": "sum"})
    .reset_index()
    .sort_values(by="Score", ascending=False)
    .head()
)

top_categories
```
